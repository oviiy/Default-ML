```python
import pandas as pd
import matplotlib as mplt
import seaborn as sns
import matplotlib. pyplot as plt 
import numpy as np
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
mortgage = pd.read_csv('mortgage.csv')
mortgage
```

```python
mortgage.shape
```

```python
mortgage.describe()
```

```python
mortgage.interpolate(method='nearest', inplace=True)
mortgage.isna().sum()
```

```python
mortgage.groupby('status_time').describe()
```

```python
mg_f = mortgage.groupby('id').first()
mg_l = mortgage.groupby('id').last()
```

```python
mg_f['default_time'].value_counts()
```

```python
mg_l['default_time'].value_counts()
```

```python
mg_f
```

```python
drops = list(mg_l.columns)[0:-3]
mg_l.drop(columns=drops, inplace=True)
```

```python
drops_f = ['time', 'orig_time', 'first_time', 'mat_time','default_time', 'payoff_time', 'status_time']
mg_f.drop(columns=drops_f, inplace=True)
```

```python
mg_ml = mg_f.join(mg_l, on='id', how='inner')
```

```python
drops_ml = ['payoff_time', 'status_time']
mg_ml.drop(columns=drops_ml, inplace=True)
```

```python
default_status_df = mg_ml['default_time']
mg_ml.drop(columns=['default_time'], inplace=True)
mg_ml['default_status'] = default_status_df
```

```python
mg_ml
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler(feature_range=(0, 1))
mg_rescaled = minmax_scaler.fit_transform(mg_ml)
mg_ml_rescaled = pd.DataFrame(mg_rescaled, columns = mg_ml.columns)
```

```python
mg_ml_rescaled.to_csv('mg_ml_rescaled.csv')
```

```python
for_hist = list(mg_ml_rescaled.columns)[:]
hist = mg_ml_rescaled[for_hist].hist(bins=20,figsize=[20,15])
```

```python
mg_ml_rescaled
```

```python
fig, ax = plt.subplots(figsize=(15,15))
mask = np.zeros_like(mg_ml_rescaled.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(mg_ml_rescaled.corr(), mask=mask, annot=True);
```
